```python
from random import choices, randint
```

```python
grade_types = {
            "final" : 45,
            "midterm" : 25,
            "assigments" : 10,
            "quiz" : 20
        }

students = [id for id in range(1,6)]
students.append(21)

courses = [11,12,13,14,15]

```

```python
query_interim = "INSERT INTO InterimGrades(GradeID,StudentID,CourseID , Type, Percentage, Val)\nVALUES\n"
query_final = "INSERT INTO FinalGrade(FinalGrade,StudentID,CourseID , Val)\nVALUES\n"

counter_interim = 1
counter_final = 1
for student in students:
    for course in courses:
        total = 0.0
        for grade_type, percetange in grade_types.items():
            grade_per_type = randint(0,100)
            total += grade_per_type * percetange/100
            query_interim += f"({counter_interim},{student},{course},'{grade_type}',{percetange},{grade_per_type}),\n"
            counter_interim += 1
        query_final += f"({counter_final},{student},{course},{round(total,2)}),\n"
            
query_interim += ";"
query_final += ";"


```

```python
with open("interim", "w") as f:
    f.write(query_interim)
```
